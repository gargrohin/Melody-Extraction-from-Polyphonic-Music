```python
root = '/mnt/data/datasets/MIR-1K/'
```

```python
singers=[]
i=-1
for file in os.listdir(root):
    if file.split('.')[1]=='wav':
        if len(singers) == 0:
            singers.append(file.split('_')[0])
            i+=1
            continue
        if file.split('_')[0] in singers:
            continue
        else:
            singers.append(file.split('_')[0])
            i+=1
    
```

```python
songs={}
```

```python
for singer in singers:
    i=0
    while(1):
        try:
            i+=1
            song = singer + '_' + str(i) + '.wav'
            file = open(root + song, 'r')
        except:
            songs[singer] = i-1
            break
```

```python
songs
```

```python
dest_bg = '/mnt/data/datasets/MIR_bg/'
dest_voc = '/mnt/data/datasets/MIR_voc/'
```

```python
for singer in singers:
    for i in range(songs[singer]):
        song = singer + '_' + str(i+1) + '.wav'
        s,sr = librosa.load(root + song, sr = 44100, mono = False)
        dest_bg_path = dest_bg + singer + '_' + str(i+1) +'_bg_' + '.wav'
        dest_voc_path = dest_voc + singer + '_' + str(i+1) +'_voc_' + '.wav'
        librosa.output.write_wav(dest_bg_path , y = s[0], sr=sr)
        librosa.output.write_wav(dest_voc_path , y = s[1], sr=sr)
        print(song)
```
