```python
import numpy as np
import librosa
import librosa.display as display
import matplotlib.pyplot as plt
import IPython
```

```python
cd ../
```

```python
sr = 22050
```

```python
WF0 = np.load('WF.npy')
WGAMMA = np.load('Wgamma.npy')
```

```python
Hgamma_vio = np.load('Hgamma_violin.npy')
```

```python
WPHI_vio = np.matmul(WGAMMA , Hgamma_vio)
```

```python
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.amplitude_to_db(WPHI_vio, ref=np.max),
                           y_axis='linear',x_axis='time', sr=sr,hop_length=256)
plt.colorbar()
plt.tight_layout()
```
